```python
import os
import sys
from datetime import datetime
sys.path.append(os.path.realpath('..'))

import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_name = "~/Downloads/openethdata_eth_data.parquet"
df = pd.read_parquet(file_name)
```

```python
df2 = df[df['date'] > datetime(2023, 3, 6)]
df2['relay'].unique()
```

```python
df2['week'] = df2.date.dt.isocalendar().week
df2['relayers'] = df2['relay']
df2.loc[(df2['relay'] != 'flashbots') & (df2['relay'] != 'ultrasound'), 'relayers'] = 'other'
df2.loc[df2['relay'] == 'bloxroute (max profit)', 'relayers'] = 'bloxroute (combined)'
df2.loc[df2['relay'] == 'bloxroute (ethical)', 'relayers'] = 'bloxroute (combined)'
df2.loc[df2['relay'] == 'bloxroute (regulated)', 'relayers'] = 'bloxroute (combined)'
df2
```

```python
df3 = df2.groupby(['week', 'relayers']).sum()['slot']

# df3.head(15)
```

```python
df4 = df3.reset_index().set_index('week').pivot(columns='relayers', values='slot')
columnsTitles = ['ultrasound', 'flashbots', 'bloxroute (combined)', 'other']


to_plot = df4.div(df4.sum(axis=1), axis=0)*100
to_plot = to_plot.reindex(columns=columnsTitles)

ax = to_plot.plot(kind='bar', stacked=True, figsize=(20, 10))
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=12)
for c in ax.containers:
    ax.bar_label(c, label_type='center', fmt='%.1f%%')
plt.axvline(x=1.5, color="red",linestyle="--",lw=2)
plt.text(1.3,75,"Ultrasound deploys" + "\n" + "Optimistic relay", fontsize=12)
ax.set_title("Percentage of weekly blocks relayed by different relays")
ax.set_ylabel("Percentage of weekly blocks relayed", fontsize=12)
ax.set_xlabel("Week of the year 2023", fontsize=12)
ax.set_xticklabels([f'Week {i}' for i in range(10, 14)], rotation=0)
```
